```python
jukjun = jukjun[['x', 'y', '장소명']]
sungbok = sungbok[['x', 'y', '장소명']]
```

```python
jukjun_demand = jukjun_demand[['cell_xcrd', 'cell_ycrd']]
jukjun_demand.columns = ['x', 'y']
sungbok_demand  = sungbok_demand[['cell_xcrd', 'cell_ycrd']]
sungbok_demand.columns = ['x', 'y']
```

```python
# 위도
latitude = 37.1778
# 경도
longitude = 127.17837
```

```python
m = folium.Map(location=[latitude, longitude],
               zoom_start=11, 
               width=2560, 
               height=1280
              ) 
```

```python
for i in range(len(jukjun)):
  print([jukjun.iloc[i, 1], jukjun.iloc[i, 0]])
  folium.Marker([jukjun.iloc[i, 1], jukjun.iloc[i, 0]], icon=folium.Icon(color = 'red'), tooltip=[jukjun.iloc[i, 2]]).add_to(m)
  folium.Circle([jukjun.iloc[i, 1], jukjun.iloc[i, 0]], radius=500).add_to(m)
```

```python
for i in range(len(sungbok)):
  print([sungbok.iloc[i, 1], sungbok.iloc[i, 0]])
  folium.Marker([sungbok.iloc[i, 1], sungbok.iloc[i, 0]], icon=folium.Icon(color = 'red'), tooltip=[sungbok.iloc[i, 2]]).add_to(m)
  folium.Circle([sungbok.iloc[i, 1], sungbok.iloc[i, 0]], radius=500).add_to(m)
```

```python
for i in range(len(jukjun_demand)):
    folium.Marker([jukjun_demand.iloc[i, 1], jukjun_demand.iloc[i, 0]], color='red', tooltip=[i, jukjun_demand.iloc[i, 1], jukjun_demand.iloc[i, 0]]).add_to(m)
```
