```python
import pandas as pd 
import numpy as np
```

```python
# 용인시 인구현항 데이터 불러오기 
path = 'C:/bigcon/input/additional_dataset/'
p_df = pd.read_csv(path + '용인시 연령별 인구 현황.csv', encoding = 'cp949')
p_df.head(10)
```

```python
import math 

# 20 ~ 60대 인구만 추출
p_df_refine = p_df.drop([0,2], axis = 0)

nm_list = p_df_refine.loc[1].tolist()
new_list = []
for i in nm_list:
    if type(i) != str :
        pass
    else:
        new_list.append(i)
print(new_list)

p_df_refine.columns = nm_list
p_df_refine_v2 = p_df_refine.loc[:,new_list].drop([1],axis=0)
p_df_refine_v3 = p_df_refine_v2[(p_df_refine_v2[' 연령별 '] == ' 20대 ') | (p_df_refine_v2[' 연령별 '] == ' 30대 ') | (p_df_refine_v2[' 연령별 '] == ' 40대 ')
              | (p_df_refine_v2[' 연령별 '] == ' 50대 ') | (p_df_refine_v2[' 연령별 '] == ' 60대 ')]
```

```python
p_df_refine_v4 = p_df_refine_v3.drop([' 용 인 시 ',' 처 인 구 ',' 기 흥 구 ',' 수 지 구 '], axis=1)
col_list = p_df_refine_v4.columns.tolist()[1:]

p_df_refine_v4 = p_df_refine_v4.apply(lambda x: x.str.strip(), axis = 1) # 앞뒤 공백 제거 
p_df_refine_v4 = p_df_refine_v4.apply(lambda x: x.str.replace(',',''), axis = 1) # 쉼표 제거
p_df_refine_v4 = p_df_refine_v4.loc[:,col_list].apply(lambda x: x.astype('float32'), axis = 1) # 실수형으로 자료형 변경 
p_df_refine_v4.loc[5] = p_df_refine_v4.loc[:,col_list].sum().tolist()
p_df_refine_v4.reset_index(drop=True, inplace=True)
p_df_refine_v4.insert(0, '연령별', ['20대','30대','40대','50대','60대','합계']) # 원하는 위치에 열 삽입 
p_df_refine_v4 = p_df_refine_v4.transpose()
p_df_refine_v4
```

```python
p_df_refine_v4.index = p_df_refine_v4.index.str.strip()
p_df_refine_v4.index = p_df_refine_v4.index.str.replace(' ','')
p_df_refine_v4.index
```
